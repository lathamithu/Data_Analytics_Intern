```
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import pandas as pd
```

```
data = pd.read_csv('sample_dataset.csv')
data.head()
```

```
i = 2
x = data.iloc[:,0]
y = data.iloc[:,1]
print (x)
print(y)
```

```
plt.scatter(x,y)
plt.ylabel('GPA',fontsize=20)
plt.xlabel('SAT',fontsize=20)
plt.show()
```

```
x1 = sm.add_constant(x)
results = sm.OLS(y,x1).fit()
results.summary()
```

```
plt.scatter(y,x)
yhat=[]
for q in y:
    yhat.append(0+1.1*q)

fig=plt.plot(yhat,x,lw=3, c='orange' , label='regressionline')
plt.show()
```
