```
import matplotlib.pyplot as plt
age = [0,10,30,50,70]
sleeping_time = [16*3600,12*3600,6*3600,7*3600,6*3600]
sal = [0,500,40000,80000,20000]
plt.scatter(age,sleeping_time,color='c',marker='*')
plt.scatter(age,sal,color='m',marker='d')
```

```
import matplotlib.pyplot as plt
year = '1999','2000','2012','2020'
adm_clg=[500,1000,1500,1200]
colors=['c','y','r','m']
plt.pie(adm_clg,labels=year,colors=colors,autopct='%.1f%%',startangle=90)
```

```
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure()
ax = Axes3D(fig)
x = [1,2,3,4,7]
y = [1,5,7,9,10]
z = [1,-1,0,1,2]
dx = np.ones(5)
dy = np.ones(5)
dz = [1,2,3,4,5]
ax.set_xlabel('x-axis')
ax.set_ylabel('y-axis')
ax.set_zlabel('z-axis')
ax.bar3d(x,y,z,dx,dy,dz,color='r')
```
